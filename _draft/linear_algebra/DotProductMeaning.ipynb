## 内積の幾何的意味

ベクトル $ \mathbf{v}_1 = (x_1, y_1) $, $ \mathbf{v}_2 = (x_2, y_2) $ について、

- ベクトルの差：
 $   \mathbf{v}_2 - \mathbf{v}_1 = (x_2 - x_1, y_2 - y_1) $

- コサイン定理を利用すると：
 $   \cos \theta = \frac{\|\mathbf{v}_1\|^2 + \|\mathbf{v}_2\|^2 - \|\mathbf{v}_2 - \mathbf{V}_1\|^2}{2 \|\mathbf{v}_1\| \|\mathbf{v}_2\|} $

- 分子の項を整理すると：
$ \|\mathbf{v}_1\|^2 + \|\mathbf{v}_2\|^2 - \|\mathbf{v}_2 - \mathbf{v}_1\|^2 = (\mathbf{v}_1 \cdot \mathbf{v}_1) + (\mathbf{v}_2 \cdot \mathbf{v}_2) - (\mathbf{v}_2 - \mathbf{v}_1) \cdot (\mathbf{v}_2 - \mathbf{v}_1) $

$ = \mathbf{v}_1 \cdot \mathbf{v}_1 + \mathbf{v}_2 \cdot \mathbf{v}_2 - (\mathbf{v}_2 \cdot \mathbf{v}_2 - \mathbf{v}_2 \cdot \mathbf{v}_1 - \mathbf{v}_1 \cdot \mathbf{v}_2 + \mathbf{v}_1 \cdot \mathbf{v}_1) $

$ = -2 \mathbf{v}_1 \cdot \mathbf{v}_2 $

- したがって、
$ \cos \theta = \frac{-2 \mathbf{v}_1 \cdot \mathbf{v}_2}{2 \|\mathbf{v}_1\| \|\mathbf{v}_2\|} = \frac{\mathbf{vV}_1 \cdot \mathbf{v}_2}{\|\mathbf{v}_1\| \|\mathbf{v}_2\|} $

- 内積の公式：
$ \mathbf{v}_1 \cdot \mathbf{v}_2 = \|\mathbf{v}_1\| \|\mathbf{v}_2\| \cos \theta $

- ベクトルの長さが内積の値に基づいて正負を持つことから、正射影を示します。

## ベクトルの射影 (Projection)

ベクトル $ \mathbf{v}_1 $ を $ \mathbf{v}_2 $ に射影する場合：

- 射影の公式：
$ \text{Proj}_{\mathbf{v}_2} \mathbf{v}_1 = \mathbf{w} $

$ \mathbf{v}_1 \cdot \mathbf{v}_2 = \|\mathbf{v}_1\| \|\mathbf{w}\| $

$ \|\mathbf{w}\| = \frac{\mathbf{v}_1 \cdot \mathbf{v}_2}{\|\mathbf{v}_2\|} $

- 射影ベクトル \( \mathbf{w} \) はスカラー倍の形式で表せます：
$ \mathbf{w} = \frac{\mathbf{v}_1 \cdot \mathbf{v}_2}{\|\mathbf{v}_2\|^2} \mathbf{v}_2 = \frac{\mathbf{v}_1 \cdot \mathbf{v}_2}{\mathbf{v}_2 \cdot \mathbf{v}_2} \mathbf{v}_2 $

- 射影の具体的な形：
$ \mathbf{w} = \frac{\mathbf{v}_1 \cdot \mathbf{v}_2}{\|\mathbf{v}_2\|^2} \mathbf{v}_2 = \frac{\mathbf{v}_1 \cdot \mathbf{v}_2}{\mathbf{v}_2 \cdot \mathbf{v}_2} \mathbf{v}_2 $
