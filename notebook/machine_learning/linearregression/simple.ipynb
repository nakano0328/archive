# 単回帰分析

このページでは、単回帰分析を行っています。

データセットの作成から分析を行っています。今回扱うデータセットは機械学習ライブラリのscikit-learnに含まれるカルフォルニアの住宅のデータセットです。

```python
#@title ライブラリのインポート

import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#@title データの読み込み
california = fetch_california_housing()
X, y = california.data, california.target
X_income = X[:, [0]]

# MedInc（Median Income）と HousePrice の散布図
plt.scatter(X_income, y, alpha=0.7, color='blue')
plt.title("Median Income vs House Price")
plt.xlabel("Median Income (MedInc)")
plt.ylabel("House Price (in $100,000)")
plt.grid(True)
plt.show()
```

```python
#@title 単回帰分析
# 単回帰分析
model = LinearRegression()
model.fit(X_income, y)

# 回帰結果
predicted_sales = model.predict(X_income)
mse = mean_squared_error(y, predicted_sales)
r2 = r2_score(y, predicted_sales)

# 結果の表示
print("回帰式: y = {:.2f} * x + {:.2f}".format(model.coef_[0], model.intercept_))
print(f"平均二乗誤差 (MSE): {mse:.2f}")
print(f"決定係数 (R^2): {r2:.2f}")

# 回帰直線のプロット
plt.scatter(X_income, y, alpha=0.7, label="Data")
plt.plot(X_income, predicted_sales, color="red", label="Regression Line")
plt.title("Advertising Cost vs Sales (Linear Regression)")
plt.xlabel("Advertising Cost")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```
